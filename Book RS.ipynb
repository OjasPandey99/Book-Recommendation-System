```python
avg_ratings = pd.DataFrame(ratings.groupby('ISBN')['BookRating'].mean())
avg_ratings['RatingCount'] = pd.DataFrame(ratings.groupby('ISBN')['BookRating'].count())
avg_ratings.sort_values('RatingCount',ascending=False).head()
```

```python
# From here we see that the most rated book does not have a good rating at all. So if we based our system solely on the 
# number of ratings, it would have been very inefficient
```

```python
# Now for ease, we are going to simplify our data by excluding users who have made less than 200 ratings and books  
# with less than 100 ratings 
```

```python
count = ratings['UserID'].value_counts()
ratings =ratings[ratings['UserID'].isin(count[count>=200].index)]
```

```python
count1 = ratings['BookRating'].value_counts()
ratings =ratings[ratings['BookRating'].isin(count1[count1>=100].index)]
```

```python
# Now we will visualize rating in a 2D array for simplicity 
```

```python
ratings = ratings.drop_duplicates(['UserID','ISBN'])  # dropping entries having same value for UserID and ISBN 
ratings_pivot = ratings.pivot('UserID','ISBN','BookRating')
```

```python
userID = ratings_pivot.index 
ISBN = ratings_pivot.columns
ratings_pivot
```

```python
print(ratings_pivot.shape)
```

```python
# Now we shall see which book is coorelated with the 2nd most rated book "LOVely Bones". From above we can see that the 
# ISBN for this book is 316666343. We will be using it. With the help of the pivot table, we can refer to a book using
# its ISBN and then we can get the rating array or rating vector of that book through that column. We'll use this column
# /vector to find correalted books using corrwith function . For correlation we wil use PearsonR coefficient. 
```

```python
bones_ratings = ratings_pivot['316666343']
similar_to_bones = ratings_pivot.corrwith(bones_ratings)
corr_bones = pd.DataFrame(similar_to_bones, columns=['pearsonR'])
corr_bones.dropna(inplace=True)
corr_summary = corr_bones.join(avg_ratings['RatingCount'])
corr_summary[corr_summary['RatingCount']>=300].sort_values('pearsonR', ascending=False).head(10)
```

```python
# we have the ISBN number but we need names of books 
```
